```python
# Create a list of prefecture codes in sorted order
codes = adjacency_dict.keys()

# Create an adjacency matrix
adj_matrix = np.zeros((len(codes), len(codes)), dtype=int)

# Populate the adjacency matrix
for i, code1 in enumerate(codes):
    for j, code2 in enumerate(codes):
        if code2 in adjacency_dict[code1]:
            adj_matrix[i, j] = 1
            adj_matrix[j, i] = 1  # Ensure the matrix is symmetric

# Create a DataFrame for the adjacency matrix
adj_df = pd.DataFrame(adj_matrix, index=codes, columns=codes)
```
