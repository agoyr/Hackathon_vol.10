```python
import os
import sys
from path_info import PROJECT_DIR, DATA_DIR
sys.path.append(PROJECT_DIR)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_adj = pd.read_csv(os.path.join(DATA_DIR, 'adjacency.csv'), index_col=0)
df_pairs = pd.read_csv(os.path.join(DATA_DIR, 'pairs_membership_by_prefecture.csv'), index_col=0)
df_pref = pd.read_csv(os.path.join(DATA_DIR, 'prefecture_data.csv'), index_col=0)

display("adjacency")
display(df_adj.head())
display()
display("pairs")
display(df_pairs.head())
display()
display("prefecture info")
display(df_pref.head())
```

```python
# 都道府県名の修正関数
def add_suffix(pref_name):
    if pref_name in ['北海道']:
        return pref_name
    elif pref_name in ['東京']:
        return pref_name + '都'
    elif pref_name in ['大阪', '京都']:
        return pref_name + '府'
    else:
        return pref_name + '県'

df_pairs_reset = df_pairs.reset_index()
df_pairs_reset['都道府県'] = df_pairs_reset['都道府県'].apply(add_suffix)
df_pairs_reset.columns = ['Prefecture_JP', 'Female_subs', 'Male_Subs', 'Total_subs']
df_pairs_reset.head()
```

```python
# Prefecture_JP 列をキーとしてマージ
df_merged = pd.merge(df_pref, df_pairs_reset, on='Prefecture_JP')
df_merged.head()
```

```python
df_merged.to_csv(os.path.join(DATA_DIR, 'prefecture_data_with_pairs_info.csv'), index=False)
```
